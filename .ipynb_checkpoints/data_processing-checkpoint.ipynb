```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Players Data Cleaning

```python
players = pd.read_csv('players.csv', index_col=[0])
```

```python
players.head()
```

```python
del players["Rk"]
```

```python
players['Player'] = players['Player'].str.replace('*', '', regex=False)
```

```python
players.head(20)
```

```python
players = players.reset_index(drop=True)
```

```python
players
```

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
# players.groupby(["Player", "Year"])
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
```

```python
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(50)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
players = players.reset_index(drop=True)
```

```python
players
```

```python
nicknames = {}
with open("nicknames.txt") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
players["Team"] = players["Tm"].map(nicknames)
```

```python
del players["Team"]
```

```python
players.insert(loc=4, column='Team', value=players["Tm"].map(nicknames))
```

```python
players
```

```python
players[players['Player'] == "Stephen Curry"][['Year','3P','3PA']]
```

```python
players.to_csv('dataset/players_preprocess.csv')
```
